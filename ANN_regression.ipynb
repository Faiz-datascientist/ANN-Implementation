```python
df_new=df.drop(['conRowNumber','CustomerId','Surname'],axis=1)
```

```python
Label_encoder=LabelEncoder()
```

```python
df_new['Gender']=Label_encoder.fit_transform(df['Gender'])
```

```python
OneHotEncoder_geo=OneHotEncoder()
```

```python
geo_arr=OneHotEncoder_geo.fit_transform(df_new[['Geography']]).toarray()
```

```python
OneHotEncoder_geo.get_feature_names_out(['Geography'])
```

```python
df_geo=pd.DataFrame(geo_arr,columns=OneHotEncoder_geo.get_feature_names_out(['Geography']))
```

```python
df_new=pd.concat([df_new.drop(['Geography'],axis=1),df_geo],axis=1)
```

```python
X=df_new.drop(['EstimatedSalary'],axis=1)
y=df['EstimatedSalary']
```

```python
y
```

```python
#Split the data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=41)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

ANN regression Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
```

```python
model=Sequential([
Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
Dense(32,activation='relu'),
Dense(1) # output layer for regression
]
)
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='mean_absolute_error',metrics='mae')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#during traning it collects logs
log_dir="Regression/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#setup early stopping 
EarlyStopping_callbacks=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
model.fit(
    X_train,y_train,validation_data=(X_train,y_train),epochs=50,callbacks=[EarlyStopping_callbacks,tensorflow_callbacks]
)
```
